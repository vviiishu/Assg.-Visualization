# Seaborn

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.random.rand(100)
y = 2 * x + 1 + np.random.normal(0, 0.1, 100)

import pandas as pd
df = pd.DataFrame({'X': x, 'Y': y})

plt.figure(figsize=(8, 6))
sns.scatterplot(x='X', y='Y', data=df)
plt.title('Scatter Plot of Two Variables')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

```python
data = np.random.normal(loc=0, scale=1, size=1000)

df = pd.DataFrame({'Value': data})

plt.figure(figsize=(8, 6))
sns.histplot(df['Value'], bins=30, kde=True)
plt.title('Distribution of a Numerical Variable')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [23, 45, 56, 78]

df = pd.DataFrame({'Category': categories, 'Value': values})

plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Value', data=df, palette='viridis')
plt.title('Comparison of Categories Based on Numerical Values')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
np.random.seed(42)
data = {
    'Category': np.random.choice(['Category A', 'Category B', 'Category C'], 300),
    'Value': np.random.normal(0, 1, 300)
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 6))
sns.boxplot(x='Category', y='Value', data=df)
plt.title('Distribution of a Numerical Variable Across Different Categories')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
np.random.seed(42)
data = np.random.multivariate_normal([0, 0, 0], [[1, 0.8, 0.5], [0.8, 1, 0.3], [0.5, 0.3, 1]], 300)

df = pd.DataFrame(data, columns=['Feature 1', 'Feature 2', 'Feature 3'])

corr_matrix = df.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}
df_3d = pd.DataFrame(data)

fig = px.scatter_3d(df_3d, x='X', y='Y', z='Z', title='3D Scatter Plot')
fig.show()
```

```python
np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}
df_violin = pd.DataFrame(data)

fig = px.violin(df_violin, x='Grade', y='Score', box=True, points='all', title='Distribution of Scores Across Grades')
fig.show()
```

```python

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_heatmap = pd.DataFrame(data)


heatmap_data = df_heatmap.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')


fig = px.imshow(heatmap_data, labels={'color': 'Sales'}, title='Variation in Sales Across Different Months and Days')
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(title='3D Surface Plot', scene=dict(
                    xaxis_title='X',
                    yaxis_title='Y',
                    zaxis_title='Z'))
fig.show()
```

```python
import plotly.express as px
import pandas as pd
import numpy as np

np.random.seed(25)

data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

fig = px.scatter(df, x='GDP', y='Population', size='Population', hover_name='Country', title='Bubble Chart of Population vs GDP',
                 labels={'GDP': 'GDP (in billions)', 'Population': 'Population (in millions)'}, size_max=60)

fig.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np
import pandas as pd

output_notebook()

np.random.seed(42)
x = np.random.rand(100)
y = np.random.rand(100)
sizes = np.random.uniform(10, 30, 100)
colors = np.random.choice(['red', 'blue', 'green', 'orange', 'purple'], 100)

df = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

p = figure(title="Random Scatter Plot", x_axis_label='x', y_axis_label='y')

p.circle('x', 'y', size='sizes', color='colors', alpha=0.6, source=df)
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource

output_notebook()

# Data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a DataFrame
data = {'fruits': fruits, 'counts': counts}
source = ColumnDataSource(data=data)

# Create a new plot
p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Counts',height=350, toolbar_location=None, tools="")

# Add a vbar renderer
p.vbar(x='fruits', top='counts', width=0.9, source=source, legend_field="fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=['red', 'orange', 'yellow', 'green'], factors=fruits))

# Show the results
show(p)
```
